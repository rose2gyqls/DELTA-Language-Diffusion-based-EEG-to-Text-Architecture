```python
import pandas as pd
from difflib import SequenceMatcher

# 1) 두 문자열 간 유사도(ratio)를 계산하는 함수 정의
def seq_similarity(a: str, b: str) -> float:
    return SequenceMatcher(None, a, b).ratio()

# 2) DataFrame에 새로운 'sim_ratio' 열 추가
df['sim_ratio'] = df.apply(lambda row: seq_similarity(row['gt'], row['predict']), axis=1)

# 3) 유사도가 높은 순으로 정렬해서 상위 몇 개를 살펴보기
df_sorted = df.sort_values(by='sim_ratio', ascending=False)

# (가) 전체에서 가장 유사도가 높은 한 행만 보고 싶다면:
best_row = df_sorted.iloc[0]
print("가장 유사한 한 쌍:")
print("GT 문장   :", best_row['gt'])
print("Predict 문장:", best_row['predict'])
print("유사도     :", best_row['sim_ratio'])

# (나) 예를 들어 상위 10개를 보고 싶으면:
top20 = df_sorted.head(20)
print("\n=== 상위 10개의 GT vs Predict (유사도 순) ===")
print(top20[['gt', 'predict', 'sim_ratio']])
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 예시: 이미 'df'라는 DataFrame에 gt, predict 열이 있다고 가정
# df = pd.read_csv('your_data.csv')  # 실제로 파일에서 불러올 때

# 1) TF-IDF 벡터라이저 학습을 위해, gt와 predict 를 합친 전체 말뭉치(corpus) 생성
corpus = df['gt'].tolist() + df['predict'].tolist()

# 2) 벡터라이저를 학습(fit)하고 변환(transform)
vectorizer = TfidfVectorizer().fit(corpus)

# 3) 각각의 열을 TF-IDF 벡터로 변환
gt_tfidf = vectorizer.transform(df['gt'])          # shape = (N_docs, N_features)
pred_tfidf = vectorizer.transform(df['predict'])   # shape = (N_docs, N_features)

# 4) 각 행(row)마다 코사인 유사도를 계산하여 리스트로 저장
cos_sim_list = []
for i in range(df.shape[0]):
    # 두 개의 1×D 벡터 간 유사도를 구해서 스칼라로 추출
    sim = cosine_similarity(gt_tfidf[i], pred_tfidf[i])[0][0]
    cos_sim_list.append(sim)

# 5) DataFrame에 'cosine_sim' 열 추가
df['cosine_sim'] = cos_sim_list

# 6) 유사도가 높은 순으로 정렬
df_sorted_tfidf = df.sort_values(by='cosine_sim', ascending=False)

# 7) 결과 출력 예시
# (가) 가장 유사한 한 쌍 보기
best_tfidf = df_sorted_tfidf.iloc[0]
print("=== TF-IDF 기반 코사인 유사도가 가장 높은 문장 쌍 ===")
print("GT 문장   :", best_tfidf['gt'])
print("Predict 문장:", best_tfidf['predict'])
print("코사인 유사도:", best_tfidf['cosine_sim'])

# (나) 상위 N개 보기 (예: N=10)
top10_tfidf = df_sorted_tfidf.head(10)
print("\n=== 상위 10개 (TF-IDF + Cosine) ===")
print(top10_tfidf[['gt', 'predict', 'cosine_sim']])
```

```python
import pandas as pd
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge_score import rouge_scorer
from bert_score import score as bertscore_score

# ===================================================
# 4) BLEU 점수 계산 함수 정의
# ===================================================
sm = SmoothingFunction().method1

def compute_bleu(reference: str, hypothesis: str, weights: tuple) -> float:
    """
    reference: 한 개의 참조 문장 (string)
    hypothesis: 예측(생성) 문장 (string)
    weights: BLEU n-그램 가중치 튜플 (예: (1,0,0,0) for BLEU-1, (0.5,0.5,0,0) for BLEU-2 등)
    """
    # Tokenize (NLTK word_tokenize 사용)
    ref_tokens = nltk.word_tokenize(reference.lower())
    hyp_tokens = nltk.word_tokenize(hypothesis.lower())
    return sentence_bleu(
        [ref_tokens],
        hyp_tokens,
        weights=weights,
        smoothing_function=sm
    )

# ===================================================
# 5) DataFrame에 BLEU-1 ~ BLEU-4 계산하여 컬럼 추가
# ===================================================
# BLEU-1: weights=(1,0,0,0)
# BLEU-2: weights=(0.5,0.5,0,0)
# BLEU-3: weights=(1/3,1/3,1/3,0)
# BLEU-4: weights=(0.25,0.25,0.25,0.25)

df["bleu1"] = df.apply(lambda row: compute_bleu(row["gt"], row["predict"], (1, 0, 0, 0)), axis=1)
df["bleu2"] = df.apply(lambda row: compute_bleu(row["gt"], row["predict"], (0.5, 0.5, 0, 0)), axis=1)
df["bleu3"] = df.apply(lambda row: compute_bleu(row["gt"], row["predict"], (1/3, 1/3, 1/3, 0)), axis=1)
df["bleu4"] = df.apply(lambda row: compute_bleu(row["gt"], row["predict"], (0.25, 0.25, 0.25, 0.25)), axis=1)

# ===================================================
# 6) ROUGE-1 점수 계산 (precision, recall, f1) 함수 정의
# ===================================================
scorer = rouge_scorer.RougeScorer(["rouge1"], use_stemmer=True)

def compute_rouge1_prf(reference: str, hypothesis: str):
    """
    rouge_scorer를 사용하여 'rouge1'의 precision, recall, fmeasure를 반환
    """
    scores = scorer.score(reference, hypothesis)["rouge1"]
    return scores.precision, scores.recall, scores.fmeasure

# DataFrame에 ROUGE-1 세 개의 컬럼(p, r, f) 추가
rouge1_p_list = []
rouge1_r_list = []
rouge1_f_list = []

for idx, row in df.iterrows():
    p, r, f = compute_rouge1_prf(row["gt"], row["predict"])
    rouge1_p_list.append(p)
    rouge1_r_list.append(r)
    rouge1_f_list.append(f)

df["rouge1_p"] = rouge1_p_list
df["rouge1_r"] = rouge1_r_list
df["rouge1_f"] = rouge1_f_list

# ===================================================
# 7) BERT-Score 계산
# ===================================================
# bert_score.score 함수는 리스트 단위로 한 번에 계산하는 것이 훨씬 빠릅니다.
# 아래에서는 전체 DataFrame의 'predict' vs 'gt'를 한 번에 계산한 뒤,
# 반환된 P, R, F 리스트를 DataFrame에 붙이는 방식으로 합니다.

# (주의) lang='en' 또는 'en-large' 등 사용 언어에 맞춰서 설정 가능. 
# 필요하다면 rescale_with_baseline=True 옵션도 붙이세요 (언어별 baseline 조정).
P, R, F = bertscore_score(
    df["predict"].tolist(),
    df["gt"].tolist(),
    lang="en",                  # 언어 설정 (예: "en", "en-large")
    rescale_with_baseline=True   # 옵션: True로 설정하면 BERT-Score를 baseline으로 보정
)

# 반환된 P, R, F는 torch.Tensor 또는 list 타입이므로 numpy로 변환하거나
# 바로 float 형태로 DataFrame에 할당해도 됩니다.
# 아래에서는 tensor → float로 변환하여 새로운 컬럼에 추가합니다.
df["bert_p"] = [float(x) for x in P]
df["bert_r"] = [float(x) for x in R]
df["bert_f"] = [float(x) for x in F]

# ===================================================
# 8) 결과 확인하기
# ===================================================
# (가) DataFrame head 출력
pd.set_option("display.max_colwidth", None)
print(df.head(10))

# (나) 전체 평균값(각 metric 별) 출력
metrics = ["bleu1", "bleu2", "bleu3", "bleu4", 
           "rouge1_p", "rouge1_r", "rouge1_f", 
           "bert_p", "bert_r", "bert_f"]

print("\n=== 전체 평균(metrics) ===")
print(df[metrics].mean())
```
